```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
survey_df = pd.read_csv(file, encoding = 'iso-8859-1', low_memory=False)
survey_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc
extract_df = survey_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extract_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract_df = extract_df.replace({0:"No", 1:"Yes"})
extract_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = extract_df.loc[extract_df["AttendedBootcamp"] == "Yes"]
attended_df.head()
```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts
bc_name = attended_df["BootcampName"].value_counts()
#bc_name_df = pd.DataFrame(bc_name)
bc_name = bc_name.reset_index()
bc_name.columns = ["BootcampName","Count"]
bc_name.head()
```

```python
# Replace 'No' and 'Yes' with '0' and '1' respectively
convert_to_num = attended_df.replace({"No":0, "Yes":1})
convert_to_num.head()
```

```python
# Group by cols -- Name, Recommend
groupby_name = convert_to_num.groupby("BootcampName")
groupby_name.sum().head()
```

```python
# extract only BootcampRecommend out
recommender = groupby_name["BootcampRecommend"]
total_rec = recommender.sum().head()
```

```python
# convert SeriesGroupby to DataFrame
recommender_df = pd.DataFrame(total_rec)
recommender_df = recommender_df.reset_index()
recommender_df.rename(columns = {"BootcampRecommend":"Recommenders"})
recommender_df.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
sub_merge = pd.merge(bc_name, recommender_df, on="BootcampName", how="right")
sub_merge.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
sub_merge["%Recommend"] = (sub_merge["BootcampRecommend"] / sub_merge["Count"]) * 100
sub_merge.head()
```

```python
# Think of what should be the numerator and the demoninator


# Sort results in descending order
sort_sub_merge = sub_merge.sort_values(["%Recommend"], ascending=False)


# Format for percentages
sort_sub_merge["%Recommend"] = sort_sub_merge["%Recommend"].map("{:,.2f}%".format)

sort_sub_merge.head()
```

```python
# Export to excel and remove index
sort_sub_merge.to_excel("output/Bootcamppart2.xlsx", index=False)
```
